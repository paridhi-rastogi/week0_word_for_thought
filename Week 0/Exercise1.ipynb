**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])


x = np.arange(0,100)
y = x*2
z = x**2

ax1.plot(x,z)
ax1.set_xlabel("x")
ax1.set_ylabel("z")
ax2.plot(x,y)
ax1.set_xlabel("X")
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax2.set_ylabel("Y")
ax2.set_title("ZOOM")

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x=np.arange(0,100)
y = 2*x
plt.subplot(1,2,1)
plt.plot(x,y,ls='--')

x=np.arange(0,100)
y=x**2
plt.subplot(1,2,2)
plt.plot(x,y,c='r')

plt.show()
```
